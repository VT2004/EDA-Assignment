Q3. What is the distribution of discounted prices vs. actual prices?


```python
df[['discounted_price', 'actual_price']].dtypes
```

```python
df['discounted_price'] = df['discounted_price'].replace({'₹': '',}, regex = True)
df['actual_price'] = df['actual_price'].replace({'₹': '',}, regex = True)
```

```python
df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')
df['actual_price'] = pd.to_numeric(df['actual_price'], errors='coerce')
```

```python
df[['discounted_price', 'actual_price']].isna().sum()
```

```python
df['discounted_price'] = df['discounted_price'].fillna(df['discounted_price'].mean())
df['actual_price'] = df['actual_price'].fillna(df['actual_price'].mean())
```

```python
#Histogram for both prices
plt.figure(figsize = (12,6))
plt.hist(df['discounted_price'], bins = 20, alpha = 0.5, label = 'Discounted Price')
plt.hist(df['actual_price'], bins = 20, alpha = 0.5, label = 'Actual Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Distribution of Discounted Prices vs. Actual Prices')
plt.legend()
plt.show()
```

```python
#scatter plot
plt.figure(figsize = (12,6))
plt.scatter(df['discounted_price'], df['actual_price'], alpha = 0.5)
plt.xlabel('Discounted Price')
plt.ylabel('Actual Price')
plt.title('Distribution of Discounted Prices vs. Actual Prices')
plt.show()
```

Q6. What are the most popular product keywords?

```python
import re
from collections import Counter
df['keywords'] = df['product_name'].str.lower().apply(lambda x: re.sub(r'[^a-z\s,]', '', x))

keywords = df['keywords'].str.split(',').explode()

keywords = keywords.str.strip()

# Count the frequency of each keyword
word_counts = Counter(keywords)

# Get the 10 most common keywords
most_common_keywords = word_counts.most_common(10)

# Display the most common keywords
print(most_common_keywords)
```

Q7. What are the most popular product reviews?

```python
df['popular_reviews'] = df['review_title'].str.lower()

popular_reviews = df['popular_reviews'].str.split(',').explode()

popular_reviews = popular_reviews.str.strip()

word_counts = Counter(popular_reviews)

most_common_reviews = word_counts.most_common(10)

print(most_common_reviews)
```

Q7. For the top 5 most popular artists, what are their most popular tracks? List the track name for each artist.

```python
# Step 1: Get the list of top 5 artists
top_5_artists = top_5_artist['Artist']

# Step 2: Filter the original dataframe for the top 5 artists
top_5_tracks = df2[df2['Artist'].isin(top_5_artists)]

# Step 3: Sort by popularity within each artist and get the most popular track for each
most_popular_tracks = top_5_tracks.sort_values(by=['Artist', 'Popularity'], ascending=[True, False])

# Step 4: For each artist, get the most popular track
top_5_tracks_for_artists = most_popular_tracks.groupby('Artist').first()[['Track Name']]

# Display the most popular track for each of the top 5 artists
print(top_5_tracks_for_artists)
```

Q9. Does the duration of tracks vary significantly across different artists? Explore this visually using a box plot or
violin plot.

```python
plt.figure(figsize=(10,6))
sns.boxplot(x = 'Artist', y='Duration (ms)', data = df2)
plt.xlabel('Artist')
plt.ylabel('Duration (ms)')
plt.title('Duration of Tracks Across Different Artists')
plt.xticks(rotation = 90, fontsize = 5)
plt.tight_layout()
plt.grid()
plt.show()
```

The duration of tracks does vary significantly across different artists. Some artists have a relatively consistent track duration (narrow box and whiskers), while others show more variability (wider box and whiskers or numerous outliers).

Some artists exhibit outliers with either very short or very long tracks, indicating unique compositions within their portfolio.

Artists with larger interquartile ranges (the height of the boxes) or longer whiskers show greater variation in track duration.
Outliers (points outside the whiskers) represent exceptional tracks with durations significantly different from the majority.

Q10. How does the distribution of track popularity vary for different artists? Visualize this using a swarm plot or a
violin plot.

```python
plt.figure(figsize=(10, 6))
sns.swarmplot(x='Artist', y='Popularity', data=df2)
plt.xlabel('Artist')
plt.ylabel('Popularity')
plt.title('Distribution of Track Popularity by Artist')
plt.xticks(rotation=90, fontsize=5)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(x='Artist', y='Popularity', data=df2)
plt.xlabel('Artist')
plt.ylabel('Popularity')
plt.title('Distribution of Track Popularity by Artist')
plt.xticks(rotation=90, fontsize=5)
plt.tight_layout()
plt.show()
```

Some artists show a wide range of track popularity, with both highly popular and less popular tracks (e.g., wide whiskers and the presence of outliers).
Other artists have a more consistent level of popularity, with narrow interquartile ranges and shorter whiskers.

Wide interquartile ranges and long whiskers suggest a diversity in track popularity, indicating both hits and less successful tracks.
Outliers signify tracks that significantly deviate from an artist's general trend (either exceptionally popular or unpopular).
